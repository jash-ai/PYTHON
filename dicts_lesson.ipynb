## Dictionaries
A dict is an ordered collection of key / value pairs.  
Create a dict with this syntax {key1: value1, key2: value2}


```python
terms = {
    "python": "a large tropical snake that kills animals for food by winding its long body tightly around them",
    "pandas": "a large black and white animal like a bear that lives in China",
    "requests": "the action of asking for something formally and politely"
}
terms
```

We can get the value of a pair by providing the key, or using the get method

```python
# With the key used in the dictionary, we can access the value associated with it.
terms['python']
```

```python
# Note that this will return an error of the key is not found in the dictionary.
terms['snake']
```

```python
# With the get method.  Note that this will return the second optional argument if the key is not found
terms.get('python', 'not found')
```

```python
terms.get('snake', 'not found')
```

We can add a key value pair to a dict

```python
terms['keras'] = 'an open-source neural network library'
terms.get('keras')
```

We can change the value of an exsisting key using the same syntax

```python
terms['keras'] = 'a high-level interface for building and training deep learning models.'
terms.get('keras')
```

We often see the data structure below - for example, returned from an API.  
A list contains several element of type dict. Each dict has the same set of keys.

```python
patient_list = [
    {"nhs_id": "123-456-7890", "name": "Fred Bloggs", "age": 42},
    {"nhs_id": "234-567-8901", "name": "Jane Doe", "age": 48},
    {"nhs_id": "345-678-9012", "name": "John Smith", "age": 31}
]
patient_list
```

```python
# Iterate through the patient list and print the name of each patient
for patient in patient_list:
    print(patient.get('name')) # better alternative to print(patient['name'])
```

```python
# We can use a list comprehension on a list of dicts to return a list of values for a specific key
[patient.get('nhs_id') for patient in patient_list]
# [patient['nhs_id'] for patient in patient_list] # alternative
```
