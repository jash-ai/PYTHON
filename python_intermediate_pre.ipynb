```python
Given a number, print “Positive”, “Negative”, or “Zero”.
number = 5 # You can change this number to test other values

if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

```python
#From a list of numbers, return the largest number.
numbers = [10, 5, 20, 15, 30, 25] # Example list of numbers

largest_number = max(numbers)

print(f"The list of numbers is: {numbers}")
print(f"The largest number in the list is: {largest_number}")
```

```python
#Access a value from a dictionary by key.
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Access a value using its key
print(f"Name: {my_dict['name']}")
print(f"Age: {my_dict['age']}")

# You can also use .get() method to safely access a key, returning None if not found
# or a default value if specified
country = my_dict.get('country', 'Unknown')
print(f"Country: {country}")

# Accessing a non-existent key directly will raise a KeyError
# Uncomment the line below to see the error:
# print(my_dict['country'])
```

```python
#Create a string and manipulate it e.g. lowercase or uppercase it.
original_string = "Hello, World! This is a Test String."

# Convert to lowercase
lowercase_string = original_string.lower()
print(f"Original string: {original_string}")
print(f"Lowercase string: {lowercase_string}")

# Convert to uppercase
uppercase_string = original_string.upper()
print(f"Uppercase string: {uppercase_string}")
```

```python
#Create a list, access elements in a list and manipulate them e.g. append and delete elements.
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

# Access elements by index
print(f"First element: {my_list[0]}")
print(f"Third element: {my_list[2]}")
print(f"Last element: {my_list[-1]}")

# Manipulate elements: Append a new element
my_list.append(60)
print(f"List after appending 60: {my_list}")

# Manipulate elements: Delete an element by index
del my_list[1] # Deletes the element at index 1 (which is 20)
print(f"List after deleting element at index 1: {my_list}")

# Manipulate elements: Modify an element
my_list[0] = 5 # Changes the first element from 10 to 5
print(f"List after modifying first element to 5: {my_list}")
```

```python
#Create a f-string and use it to print out a variable.
name = "Colab User"
age = 30

# Using an f-string to embed variables
message = f"Hello, {name}! You are {age} years old."
print(message)

# You can also embed expressions directly
print(f"Next year, {name} will be {age + 1} years old.")
```
