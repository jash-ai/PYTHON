### Find the year that italy has the highest increase of gdb per capita

- Calculate the yearly percentage increase compared to previous year
- find the year that has highest increase in terms of percentage.

```python
italy
```

calculate the yearly change in percentage

```python
#percentage change with previous number/value eg 1962 value to 1961 value
italy_pct_change=italy.pct_change()
```

```python
# multiply italy_pct_change by 100
italy_pct_change=italy_pct_change*100
```

```python
##1960 bo previous value to compare it to therefore NaN not a number
italy_pct_change
```

find the year that has highest change

```python
# idxmax() = returns a Series with the index of the maximum value for each column
# ##year of increase
italy_pct_change.idxmax()
```

```python
##amount of increase
italy_pct_change.max()
```

```python
#shows both the index and value
italy_pct_change.idxmax(),italy_pct_change.max()
```
