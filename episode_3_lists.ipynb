# **8. Lists in python:**

You will often want to work with many data points in Python, and the simplest way to do so is by creating a list.

**Introduction to lists**

You can store a lot of information in a Python list instead of having to use individual variables You create a list of data using square barckets and commas in python. 

Imagine you want to store the heights of four family members:

```python
#Invidual variables vs. a list. 
height_a = 1.76
height_b = 1.91
height_c = 1.61
height_d = 1.73

print(height_a, height_b, height_c, height_d)

heights = [1.76, 1.91, 1.61, 1.73]
print(heights)
```

Using a list is a significantly more efficient way of storing data in python in comparison to storing data points as individual variables. **A list is a way to give a single name to a collection of values**. The values in a list can be a variable, integers, floats, bools, or strings - and a combination thereof. 

```python
#Mixing data types in a list

variable_a = "I thought python was a snake"

#you have int, float, boolean, string in a list
data_mix = [5, 5.6, True, "Lists", variable_a]
print(data_mix)

print(type(data_mix))

#Could someone tell me what each of thse data types are?
```

**Subsetting lists**

Subsetting a python list allows you to extract information from the list

**indexing:** One way which can be used to do so is through a list index. The first element in the list has index 0 (called 0 indexing), the second element has index 1 and so on. In order to select a specifc value from a list you use the list name, square brackets and the index of the value you are intrested in. For example: `list_1[3]`

```python
#Task 23: Select the value 1948 from list_2

list_2 = ["NHSX", 2019, "NHSE/1", 1948, "PHE", 2013, "Public Health Wales", 2009]

#as this is the 4 record but as the index starts with 0,1,2,3
# index position 3 is where the value of 1948 is.

print(list_2[3])
```

Python also has negative indexes, which allows you to count backwards. This is particularly important if you want select elements at the end of your list. For example, to select 2009, the date which Public Health Wales was founded on you could use the index -1, or the index 6. All list elements have both a 0 index and a negative index. 

```python
#Task 24: Select the string PHE from list_2 using both the 0-index and the negative-index. 
print(list_2[4])

#the last value is index position -1
#negative index
print(list_2[-4])
```

**slicing:** List slicing allows you to select multiple elements from a list, and by doing so creates a new list. This is done by specifying an index range, using a colon. For example: `list_1[3:6]`

*Important note:* The index you specify before the colon, so where the slice starts, is included, while the index you specify after the colon, where the slice ends, is not. For the above example the value with index 6 would **NOT** be included in the slice

```python
#Task 25: Select all the NHS related data from list_2 using slicing
print(list_2)
list_2[:]
```

You can also choose to just leave out the index before or after the colon. For example: `list_1[3:]`. If you leave out the index where the slice should end, you include all elements up to and including the last element in the list. The same holds true if you leave out the index where the slice should start. 

```python
#Example of leaving out the index where the slice should start 
print(list_2)

print(list_2[:4])

#Returns the same as list_2[0:4]
print(list_2[0:4])
```

# **9. Loops in python:**

Using loops in python allows a program to repeat a specific beahviour and re-run a line(s) of code. 

**for Loops:**

`for` loops are generally used when you want a process to repeat *for* a fixed number of times. 

The `for` loop always requires the following structure:

Untitled-7.png
The `variable_name` can be anything as long as it isn't another one of your variable names.

The `collection` is any element that can be iterated over. For example, lists, strings, tuples, dictionaries (the keys), and sets (in no particular order). These will be covered in more detial in a subsequent section. 

**Important note:** Notice how a colon always indicates that at least the next line of code will be indented.

By creating a loop you allo a program to repeat a conditional process. A loop will continue to execute as long as a certain condition is met. There are two types of key loops in python: `for` and `while` loops.

```python
#for loop in action.

total = 0                      #Line 1 

for num in [0, 1, 2, 3, 4, 5]: #Line 2
  total = total + num          #Line 3

print(total)                   #Line 4
```

**what just happened?**

In this example, `prime_number` is an arbitrarily variable — it could just as easily be `prime` or even` x`. Each time we loop, this variable references the next element in the list `primes` until none are left.

When Line 2 is evaluated, `prime_number` is set to the first element of `primes: primes[0]`. Effectively, it sets prime_`number = primes[0]` (**Important note:** index `0` is the number `2` in this list, index `1` is the number `3`. etc.). Because Line 3 is indented, it's evaluated next and prime_number (i.e., `primes[0]` equivalent to the number `2`) is displayed.

Now, we loop back to Line 2. Here, `prime_number` is set to the next element of `primes: primes[1]`. On Line 3, we display `prime_number` again.

This continues until we loop back to Line 2 and we've iterated through each element in the `primes` list. At this point, the `for` loop is done and we jump to the next line after the code block, which is Line 4.

```python
#for loop in action #2.

primes = [2, 3, 5, 7, 11]     #Line 1        

for prime_number in primes:   #Line 2
  print(prime_number)         #Line 3

print('Done!')                #Line 4
```

To start, the variable `total `is `0`.

In Line 2, the variable `num` is set to 0, which is the first element in the list.

In Line 3, `total = total + num`, so `total = 0 + 0`. 

Therefore at the end of the first loop of the code block, `total `is still equal to `0`.

We loop back to Line 2. Now, `num` is set to the second element, `1`. In Line 3, `total = total + num`, `so total = 0 + 1`.

Once the `for` loop is done, we see that `total` is `15` (`0` + `1` + `2` + `3` + `4` + `5`), as expected.

**while Loops:**

The `while` loop A while loop is generally used in Python when we don't know when the looping will stop or how many iterations the loop will require. It is simpler and cleaner than the `for` loop. The `while` loop requires the following structure: 

Untitled-8.png

As long as the conditional is true in a `while `loop, the loop will keep going forever



```python
#A while loop in action:

counter = 0                                                       # Line 1

while counter < 10:                                               # Line 2
  print('counter is currently', counter)                          # Line 3
  counter = counter + 1                                           # Line 4

print("We're out of the loop. The value of counter is", counter)
```

Here, we're creating a while loop using the `counter` variable.

On Line 1, we're starting `counter` off at `0`.

We then evaluate if `counter` is less than `10` (Line 2). Because `0` is less than `10`, Lines 3 and 4 are run.

Line 3: We simply print the current value of counter.
Line 4: We then increment the value of counter by `1` each loop. 
Once counter is equal to `10`, the loop will break, as `10` is not less than `10`.
