# **6. Comparison operators and statements:** 

In Python, comparison operators compare the values on either sides of them and decide the relation among them. They are also called relational operators.

To tell a computer how to "decide" what to do, **comparison statements** are used. Comparison statements always boil down to a single boolean value: `True` or `False`. Remember that booleans (`bools`) are one of the four key data types in Python and represent logical values. `bools` can only be `True` or `False`.

To write comparison statements in programming *comparison operators* are used. A list of comparison operators are listed below: 

|Operator |Functionality |Example |
|-|-|-|
|`>`|Greater than| 5 > 2 |
|`>=`|Greater than or equal to|5 >= 5|
|`<`|Less than|2 < 200|
|`<=`|Less than or equal to|1 <= 72|
|`==`|Tests if the left and right sides have equal values|5 == 5|
|`!=`|Tests if the left and right sides have unequal values|3 != 5|




```python
#Comparison operators in action.
a = 17
b = 5
c = 'hello'

print(5 > 2)
print(5 < 2)
print(14 <= 14)
print(7 == 8)
print(a != b)
print(c == 'hello')
```

**Combining Conditional Statements:**

Conditional statements are a powerful tool, but they can be made to be even more powerful, by combining one or more conditional statements using *logical operators*.

By combining conditional statements, you are asking the computer to evaluate more than one condition at the same time. 

*For examples:*

1.   If it's going to be lower than 10ºC **and** raining, wear a raincoat.
2.   If it's going to be higher than 10ºC **and** raining, bring an umbrella.
3.   If it's going to be lower than 10ºC **or** windy, wear a sweater.

There are three logical operators in python, they are listed below:

|Logical Operator |Functionality |Example |
|-|-|-|
|`and`|	Evaluates to `true` if **both** statements are `true`|(`x > 10 and x < 20`)|
|`or`|	Evaluates to `true` if **one** of the statements is `true`|(`x > 10 or x < 20`)|
|`not`|	Evaluates to `true` if the result is **false** |`not (x > 10 and x < 20)`|

```python
#Combining conditional statements in action:
e = 10
f = 14


if (e and f) < 15:
  print("Good math skills")

if (e or f) < 11:
  print("Good math skills")

if not (e and f < 7):
   print("Good math skills")
```

**The elif condition in if statements:**

If you need a conditional to check for more than two possibilities, you can add more comparisons to create an if ... else if ... else chain.This is called an elif statement.

elif statements tell Python, "If something is true, do this. If not, check the next condition in the chain (etc.)

The first conditional statment that is true is executed, while the rest of the chain is skipped entirely. The syntax of an elif chain is shown below:

Untitled-6.png

`elif` statements always have one `if`, as many `elifs` as you want, and zero to one `else`. Remember that in an `elif` chain, once the computer finds a true condition, it skips the rest of the chain.

```python
#elif chain at work: 

grade = 42

if grade >= 90:
    print("You got an A.")
elif grade >= 80:
    print("You got a B.")
elif grade >= 70:
    print("You got a C.")
elif grade >= 60:
    print("You got a D.")
else:
    print("You failed, sorry.")
```
