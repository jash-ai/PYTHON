Given a temperature reading, return "cool" if under 15, "warm" if 15 or over but under 25 and "hot" if 25 or over.

```python
temperature = int(input("Enter the temperature. Choose a whole number between 0 and 30."))
# Write your code here
if temperature < 15:
    print("cool")
elif temperature < 25:
    print("warm")
else:
    print("hot")
```

Loop with for

Print the squares of whole numbers from 1 to 10

```python
for num in range(1,11):
  square = num ** 2
  print (f"the square of {num} is {square}")
```
