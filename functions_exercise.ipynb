Write a function, named rectangle_area, to return the area or a rectangle given its length and width

```python
def rectangle_area(length, width):
    return length * width
```

```python
length = 5
width = 4
# Invoke and test the function rectangle_area here
area = rectangle_area(length, width)
print(f"The area of the rectangle with length {length} and width {width} is: {area}")
```

Write a function, named rectangle_summary, to return both the area and the perimeter (think tuples) given its length and width

```python
def rectangle_summary(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter
```

```python
length = 5
width = 4
# Invoke and test the function rectangle_summary here
area_summary, perimeter_summary = rectangle_summary(length, width)
print(f"For a rectangle with length {length} and width {width}:")
print(f"  Area: {area_summary}")
print(f"  Perimeter: {perimeter_summary}")
```
